```python
def pearson_corrlation(x,y):
    n = len(x);
    numerator = np.dot(x,y) - n  * x.mean() * y.mean()
    denominator = np.sqrt( ( (x**2).sum() - n * (x.mean()**2) ) * ( (y**2).sum() - n * (y.mean() ** 2) ) )
    return numerator/denominator
```

```python
def pearson_corrlation2(x,y):
    n = len(x)
    numerator = np.dot(x,y) - n  * x.mean() * y.mean()
    denominator = n * x.std() * y.std()
    return numerator/denominator
```

```python
pearson_corrlation(x,y)
```

```python
pearson_corrlation2(x,y)
```

```python
np.corrcoef(x,y)
```

```python
#Sklearn LinearRegression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x.reshape(-1,1), y.reshape(-1,1))
print('Slope: \n', lr.coef_)
print('Intercept: \n', lr.intercept_)
```

```python
line_x = np.linspace(0, 2., 100)
plt.scatter(x, y)
plt.grid(True)
plt.plot(line_x, lr.intercept_ + lr.coef_[0]*line_x, c='red')
plt.show()
```

```python
def MSE(y,y_hat):
    n = len(y)
    mse =  (1/n) * ((y_hat-y)**2).sum()
    return np.round(mse,2)
```

```python
x = np.round(np.random.random((500,1)),3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
print('Slope: \n', lr.coef_)
print('Intercept: \n', lr.intercept_)
```

```python
y_hat = lr.predict(X_test)
```

```python
MSE(y_hat,y_test)
```

```python
line_x = np.linspace(0,1,100 )
plt.scatter(X_test, y_test)
plt.grid(True)
plt.plot(line_x, lr.intercept_ + lr.coef_[0]*line_x, c='green',lw=2)
plt.show()
```
